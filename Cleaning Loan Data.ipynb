```python
# Load the dataset into pandas
import pandas as pd

from google.colab import files
uploaded = files.upload()

df = pd.read_csv('bank_loan.csv')

# Inspect the first few rows of the DataFrame
print("First five rows of the dataset:")
print(df.head())

# Get a summary of the DataFrame
print("\nDataFrame summary:")
print(df.info())
```

```python
# Step 1: Detect and fill missing values
df['CCAvg'] = df['CCAvg'].astype(str)

# Convert 'CCAvg' values like '1/60' to float
def convert_ccavg(value):
    if '/' in value:
        parts = value.split('/')
        return float(parts[0]) + float(parts[1]) / 100 if parts[1] != '00' else float(parts[0])
    try:
        return float(value)
    except:
        return None

df['CCAvg'] = df['CCAvg'].apply(convert_ccavg)

# Fill missing values
df_filled = df.fillna({
    'CCAvg': df['CCAvg'].mean(),
    'Experience': df['Experience'].mean(),
    'Income': df['Income'].mean(),
    'ZIP Code': 'Unknown'
}, inplace=True)
print("DataFrame after filling missing values:\n", df_filled)

# Drop rows with any missing values
df_dropped = df.dropna()
print("DataFrame after dropping rows with missing values:\n", df_dropped)

# Step 2: Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("DataFrame after removing duplicates:\n", df_no_duplicates)

# Step 3: Convert columns to appropriate types
df['ZIP Code'] = df['ZIP Code'].astype(str)
df['ID'] = df['ID'].astype(str)
print("DataFrame after converting types:\n", df)

# --- Step 4: Rename column for clarity ---
df.rename(columns={'CCAvg': 'CreditCardAvg'}, inplace=True)

# Show cleaned data preview
df.head()
```
