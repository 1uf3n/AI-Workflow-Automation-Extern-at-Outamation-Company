```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/bank_loan.csv')
print(df.head())
```

```python
# 1. Clean and convert 'CCAvg' from strings to floats
def convert_ccavg(value):
    try:
        return float(value.replace('/', '.'))
    except:
        return None  # for malformed entries

df['CCAvg'] = df['CCAvg'].apply(convert_ccavg)

# 2. Remove rows with missing or unparseable 'CCAvg' entries
df = df.dropna(subset=['CCAvg'])

# 3. Remove duplicates if any (based on all columns)
df = df.drop_duplicates()

# 4. Remove invalid rows: e.g., negative age or experience
df = df[(df['Age'] >= 0) & (df['Experience'] >= 0)]

# 5. Drop columns that are not useful for modeling
df_cleaned = df.drop(columns=['ID', 'ZIP Code'])

# 6. Reset index after cleaning
df_cleaned.reset_index(drop=True, inplace=True)

cleaned_file_path = 'bank_loan_cleaned.csv'
df_cleaned.to_csv(cleaned_file_path, index=False)
```
